### Create a bar chart showing various demographic breakdowns for Berkeley (race or gender)

```python
uc_berkeley_data = df[df['name'] == 'University of California-Berkeley'].iloc[0]

demographic_columns = [
    'men', 'women', 'another_gender',
    'american_indian_or_alaska_native', 'asian', 'black', 'hispanic',
    'native_hawaiian_or_pacific_islander', 'white', 'two_or_more', 'unknown', 'nonresident'
]

uc_berkeley_demographics = uc_berkeley_data[demographic_columns]

plt.figure(figsize=(14, 7))
sns.barplot(x=uc_berkeley_demographics.index, y=uc_berkeley_demographics.values, palette='viridis')
plt.title('Demographic Breakdown for University of California-Berkeley')
plt.xlabel('Demographic Category')
plt.ylabel('Number of Students')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Create several charts showing some interesting elements you find in these data

Some ideas might be comparing UC Berkeley to other UCs, looking at Ivy League comparisons, running some regressions

```python
# Get data for UC Berkeley and UCLA
berkeley_ucla_data = df[df['name'].isin(['University of California-Berkeley', 'University of California-Los Angeles'])]

# Select gender columns and college name
gender_comparison = berkeley_ucla_data[['name', 'men', 'women', 'another_gender']].set_index('name')

# Reshape data for plotting with seaborn
gender_comparison_melted = gender_comparison.reset_index().melt(id_vars='name', var_name='Gender Category', value_name='Number of Students')

plt.figure(figsize=(12, 7))
sns.barplot(x='name', y='Number of Students', hue='Gender Category', data=gender_comparison_melted, palette='tab10')
plt.title('Gender Distribution: UC Berkeley vs. UCLA')
plt.xlabel('University')
plt.ylabel('Number of Students')
plt.xticks(rotation=0)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()

# Prepare data for racial/ethnic comparison
racial_demographic_columns = [
    'american_indian_or_alaska_native', 'asian', 'black', 'hispanic',
    'native_hawaiian_or_pacific_islander', 'white', 'two_or_more', 'unknown', 'nonresident'
]

race_comparison = berkeley_ucla_data[['name'] + racial_demographic_columns].set_index('name')

race_comparison_melted = race_comparison.reset_index().melt(id_vars='name', var_name='Racial Category', value_name='Number of Students')

plt.figure(figsize=(15, 8))
sns.barplot(x='name', y='Number of Students', hue='Racial Category', data=race_comparison_melted, palette='tab20')
plt.title('Racial/Ethnic Demographic Breakdown: UC Berkeley vs. UCLA')
plt.xlabel('University')
plt.ylabel('Number of Students')
plt.xticks(rotation=0)
plt.legend(title='Race/Ethnicity', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Percentage of Women: UC Berkeley vs. other UC schools

```python
# Filter for University of California schools
uc_schools = df[df['name'].str.contains('University of California-', na=False)].copy()

# Calculate percent_women for each UC school
uc_schools['percent_women'] = (uc_schools['women'] / uc_schools['enrollment']) * 100

# Sort by percent_women for better visualization if desired, or keep original order
uc_schools_sorted = uc_schools.sort_values(by='percent_women', ascending=False)

plt.figure(figsize=(14, 7))
sns.barplot(x='name', y='percent_women', data=uc_schools_sorted, palette='viridis')
plt.title('Percentage of Women Enrollment at University of California Schools')
plt.xlabel('University of California Campus')
plt.ylabel('Percentage of Women (%)')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 100) # Percentage should be between 0 and 100
plt.tight_layout()
plt.show()
```

### Percentage of Another Gender Students: UC Berkeley vs. other UC schools

```python
# Filter for University of California schools (already done in previous step, reusing uc_schools)
# uc_schools = df[df['name'].str.contains('University of California-', na=False)].copy()

# Calculate percent_another_gender for each UC school
uc_schools['percent_another_gender'] = (uc_schools['another_gender'] / uc_schools['enrollment']) * 100

# Sort by percent_another_gender for better visualization if desired, or keep original order
uc_schools_another_gender_sorted = uc_schools.sort_values(by='percent_another_gender', ascending=False)

plt.figure(figsize=(14, 7))
sns.barplot(x='name', y='percent_another_gender', data=uc_schools_another_gender_sorted, palette='plasma')
plt.title('Percentage of Another Gender Enrollment at University of California Schools')
plt.xlabel('University of California Campus')
plt.ylabel('Percentage of Another Gender (%)')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, uc_schools_another_gender_sorted['percent_another_gender'].max() * 1.1) # Set y-limit dynamically
plt.tight_layout()
plt.show()
```
